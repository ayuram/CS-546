```python
import random
from datasets import load_dataset

# Define the datasets
datasets = {
    "tau/commonsense_qa": load_dataset("tau/commonsense_qa", trust_remote_code=True)['train'],
    # "Idavidrein/gpqa": load_dataset("Idavidrein/gpqa", "gpqa_main", trust_remote_code=True)['train'],
    # "BAAI/TACO": load_dataset("BAAI/TACO", trust_remote_code=True)['test'],
    "b-mc2/sql-create-context": load_dataset("b-mc2/sql-create-context", trust_remote_code=True)['train'],
    "basicv8vc/SimpleQA": load_dataset("basicv8vc/SimpleQA", trust_remote_code=True)['test'],
    "nvidia/OpenMathInstruct-2": load_dataset("nvidia/OpenMathInstruct-2", trust_remote_code=True)['train'],
}

dataset_names = list(datasets.keys())

keys = {
    'tau/commonsense_qa': [('question', 'choices'), ('answerKey', None)],
    'Idavidrein/gpqa': [('question', None), ('choices', None)],
    'BAAI/TACO': [('question', None), ('solutions', None)],
    'b-mc2/sql-create-context': [('context', 'question'), ('answer', None)],
    'basicv8vc/SimpleQA': [('problem', None), ('answer', None)],
    'nvidia/OpenMathInstruct-2': [('problem', None), ('generated_solution', 'expected_answer')],
}

def get_random_qa(dataset_name):
    dataset = datasets[dataset_name]
    sample = random.choice(dataset)
    question_keys, answer_keys = keys[dataset_name]
    question = '\n'.join([str(sample[key]) for key in question_keys if key])
    answer = '\n'.join([str(sample[key]) for key in answer_keys if key])
    return question, answer

from tqdm import tqdm
# Generate MEGA QUESTIONS
mega_questions = []
for i in tqdm(range(50)):
    n = 2
    selected_datasets = random.sample(dataset_names, n)
    
    mega_question_parts = []
    for idx, dataset_name in enumerate(selected_datasets):
        question, answer = get_random_qa(dataset_name)
        # if question and answer:
        mega_question_parts.append(f"## Question {idx + 1}\n{question}\n\n## Answer {idx + 1}\n{answer}")
    
    mega_question = f"\n\n".join(mega_question_parts)
    mega_questions.append(mega_question)

# Print or save the MEGA QUESTIONS
for mq in mega_questions:
    print(mq)
    print("---")
```

```python
import pandas as pd
df = pd.DataFrame(mega_questions, columns=['Question'])
df.to_csv('mega_questions_samples.csv', index=False)
print(df.head())
```
